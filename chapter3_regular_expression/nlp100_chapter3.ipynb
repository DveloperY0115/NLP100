```python
# Problem 20
import json
import os

json_documents = {}

with open('enwiki-country.json', 'r', encoding='UTF-8') as f:
    for line in f:
        json_dict = json.loads(line)
        title = json_dict['title']
        body = json_dict['text']
        json_documents[title] = body
        
uk_article = json_documents['United Kingdom']

print(uk_article)
```

```python
# Problem 21
import re
p = re.compile('\[\[Category:.*\]\]')

category_lines = p.findall(uk_article)
print(category_lines)
```

```python
# Problem 22

for i in range(len(category_lines)):
    line = category_lines[i]
    line = line.replace('[[', '')
    line = line.replace('Category:', '')
    line = line.replace(']]', '')
    category_lines[i] = line
print(category_lines)
```

```python
# Problem 23
import re
p = re.compile('==[^=]+==')   # [^=]+ means any string without = inside

section_headers = p.findall(uk_article)

for i in range(len(section_headers)):
    line = section_headers[i]
    line = re.sub('=*', '', line)
    section_headers[i] = line
print(section_headers)
```

```python
# Problem 24

p = re.compile('<ref>.*</ref>')

refs = p.findall(uk_article)

# print(refs)

urls = []

for line in refs:
    url_pattern = re.compile('url=.* \|')
    url = url_pattern.findall(line)
    if url:
        urls.append(url)
    
print(urls)
```

```python
# Problem 25

p = re.compile('{{Infobox.*}}', re.DOTALL)

infobox_txt = p.findall(uk_article)[0]
infobox_txt = infobox_txt.replace('\n', '')
infobox_txt = infobox_txt.replace(' ', '')

p = re.compile('^|.*_name=.*|$')
name = p.search(infobox_txt)
print(name)
```
